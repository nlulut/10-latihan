# Menambahkan Library

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

# Membuat Variabel Gambar

```
car = cv2.imread('/content/drive/MyDrive/fotografi /Tugas/Gambar (bahan)/car.jpg')
burung = cv2.imread('/content/drive/MyDrive/fotografi /Tugas/Gambar (bahan)/burung.jpeg')
lumba = cv2.imread('/content/drive/MyDrive/fotografi /Tugas/Gambar (bahan)/lumba.jpg')
```

# Menampilkan Gambar

```
ukuran_baru = (300, 200) # Lebar, Tinggi

car_resized = cv2.resize(car, ukuran_baru) #Mengecilkan ukuran gambar
burung_resized = cv2.resize(burung, ukuran_baru) #Mengecilkan ukuran gambar
lumba_resized = cv2.resize(lumba, ukuran_baru) #Mengecilkan ukuran gambar

cv2_imshow(car_resized)
cv2_imshow(burung_resized)
cv2_imshow(lumba_resized)
```

# Menampilkan Dimensi

```
print("Car")
print(car.shape)
print("Burung")
print(burung.shape)
print("Lumba")
print(lumba.shape)
```

Konversi Warna

```
burungabu=cv2.cvtColor(burung,cv2.COLOR_BGR2GRAY)
burungrgb=cv2.cvtColor(burung,cv2.COLOR_BGR2RGB)
burunghsv=cv2.cvtColor(burung,cv2.COLOR_BGR2HSV)
```

# Menampilkan Gambar Konversi

```
ukuran_baru = (300, 200) #lebar, tinggi

burungabu_resized = cv2.resize(burungabu, ukuran_baru) #Mengecilkan ukuran gambar
burungrgb_resized = cv2.resize(burungrgb, ukuran_baru) #Mengecilkan ukuran gambar
burunghsv_resized = cv2.resize(burunghsv, ukuran_baru) #Mengecilkan ukuran gambar

cv2_imshow(burungabu_resized)
print(burungabu_resized.shape)
print(burungabu_resized.size)
print(burungabu_resized.dtype)

cv2_imshow(burungrgb_resized)
print(burungrgb_resized.shape)
print(burungrgb_resized.size)
print(burungrgb_resized.dtype)

cv2_imshow(burunghsv_resized)
print(burunghsv_resized.shape)
print(burunghsv_resized.size)
print(burunghsv_resized.dtype)
```

Memisah chanel warna

```
B,G,R=cv2.split(burung)
ukuran_baru = (300, 200) # Lebar, Tinggi

B_resized = cv2.resize(B, ukuran_baru) #Mengecilkan ukuran gambar
G_resized = cv2.resize(G, ukuran_baru) #Mengecilkan ukuran gambar
R_resized = cv2.resize(R, ukuran_baru) #Mengecilkan ukuran gambar

print("Blue")
cv2_imshow(B_resized)
print(B_resized.shape)
print("Green")
cv2_imshow(G_resized)
print(G_resized.shape)
print("Red")
cv2_imshow(R_resized)
print(R_resized.shape)
```

Menggabungkan Chanel

```
B,G,R=cv2.split(burung)
ukuran_baru = (300, 200) # Lebar, Tinggi

B_resized_temp = cv2.resize(B, ukuran_baru) #Mengecilkan ukuran gambar
G_resized_temp = cv2.resize(G, ukuran_baru) #Mengecilkan ukuran gambar
R_resized_temp = cv2.resize(R, ukuran_baru) #Mengecilkan ukuran gambar

gabungrgb = cv2.merge((R_resized_temp,G_resized_temp,B_resized_temp))
cv2_imshow(gabungrgb)

gabungbgr = cv2.merge((B_resized_temp,G_resized_temp,R_resized_temp))
cv2_imshow(gabungbgr)
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
